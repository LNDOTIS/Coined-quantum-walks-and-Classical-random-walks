# Coined quantum walk on a hypercube

***Note:*** This implementation is adapted from the quantum walk models available at: https://github.com/hiperwalk/hiperwalk/tree/master/examples/coined

```python
import hiperwalk as hpw
import numpy as np
import matplotlib.pyplot as plt
```

The dimension of hypercube we want to create.

```python
n = 5
hypercube = hpw.Hypercube(n)
```

Create a quantum walk on hypercube, which is an instance of the Coined class. 

```python
qw = hpw.Coined(hypercube)
```

Initializing the state at the center.

```python
ini_state = qw.state([(1/np.sqrt(n), (0,2**i)) for i in range(n)])
print(ini_state)
```

### Probability distribution over the hypercube graph

```python
final_state = qw.simulate(time=100, initial_state=ini_state)
probability = qw.probability_distribution(final_state)
hpw.plot_probability_distribution(probability, graph=hypercube, cmap='winter', filename = 'probability_distribution_cube')
```

### Bar graph of probability distribution

```python
hpw.plot_probability_distribution(probability, plot='bar', filename = 'final_distribution')
```
